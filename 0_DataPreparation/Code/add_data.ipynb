# Hinzufügen von zusätzlichen Daten

1.) Feiertage in Schleswig-Holstein

```python
# Read data (umsatz_gesamt.csv)
df_umsatz = pd.read_csv('../Data/umsatz_gesamt.csv', parse_dates=['Datum'])

# Read Feiertage.csv
df_feiertage = pd.read_csv('../Data/Feiertage.csv', sep=';', parse_dates=['Datum'])
df_feiertage['Datum'] = pd.to_datetime(df_feiertage['Datum'])

# create holiday set
feiertage_set = set(df_feiertage['Datum'].dt.date)

# find sundays
def is_sunday(date):
    return date.weekday() == 6

# code sundays and holidays
def is_holiday_or_sunday(date):
    return is_sunday(date) or date.date() in feiertage_set

# new column sun- or holiday
df_umsatz['sunday_or_holiday'] = df_umsatz['Datum'].apply(is_holiday_or_sunday).astype(int)

df_umsatz
```

```python
# save to csv
df_umsatz.to_csv('../Data/umsatz_gesamt_feiertage.csv', index=False)
```
